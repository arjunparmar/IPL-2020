```python
X_train = X.iloc[:17290, :]
X_test = X.iloc[17289:, :]
```

```python
Y_train = Y.iloc[:17290, :]
Y_test = Y.iloc[17289:, :]
```

```python
def biasing_X(X):
    N = X.shape
    arr = np.ones((N[0], 1))
    X = np.hstack((arr, X))
    w = np.zeros((N[1]+1, 1))
    return X, w
```

```python
X_bias, theta = biasing_X(X_train)
```

```python

X_bias = np.array(X_bias)
X_bias = X_bias.T
print(X_bias.shape)
```

```python
theta = np.array(theta)
theta.shape
```

```python
def hypo(X, W):
    return np.dot(X.T, W)
```

```python
def cost(X, Y, W):
    m = X.shape[1]
    #print(m)
    n = X.shape[0]
    sum = 0
    Temp = hypo(X, W) - Y
    J = (np.sum(Temp**2))/(2*m)
    #print(J)
    return J
```

```python
def gradient_descent(X, W, Y, Alpha, Step):
    N = X.shape[1]
    print(N)
    J = []
    initial = cost(X, Y, W)
    J.append(initial)
    #np.dot(x.T, (y_hat-y)))
    print(J)
    for i in range(Step):
        Te = hypo(X, W)
        Te = Te - Y
        K = X@Te
        print(i)
        #print(K.shape)
        #print(K.T)
        W = W - (Alpha/N)*K
        Temp = cost(X, Y, W)
        J.append(Temp)
        
    return J, W
```

```python
J, W = gradient_descent(X_bias, theta, Y_train, 0.009, 1000)
```

```python
print(W) # value from Gradient Descent
```

```python
theta1 = np.linalg.pinv(X_bias@X_bias.T)@(X_bias@Y_train)
theta1 #value from formula
```

```python
def prediction(W1):
    X_b1, _ = biasing_X(X_test)
    Y_pred = X_b1@W1
    return Y_pred
```

```python
y_bar = np.mean(Y_test)
y_bar
```

```python
def error(Y_pp):
    R_error = ((Y_pp - y_bar).T@(Y_pp - y_bar)) /((Y_test - y_bar).T@(Y_test- y_bar))
    return R_error
```

```python
Y_pred1 = prediction(W)
error(Y_pred1) # predicted by descent
```
